```python
CFdictionary = {'lat': np.array(da['lat']), 'lon': np.array(da['lon']),'ID':np.array(da['ID'])}
```

```python
HGT=["50","80","100","150"]
NameList2=[]

for i in HGT:
    NameDir=list(map(lambda x: x+"-"+i,np.array(da['Turb'])))
    NameList2.extend(NameDir)
    
print(len(NameList2))
print()
CFList=(list(map(lambda x:'Capacity Factor [%]-'+x,NameList2)))

CFData=[]
for i in np.array(da['hgt']):
    for j in np.array(da['Turb']):  
        CFData.append(np.array(da['CF'].sel(hgt=i,Turb=j)))


print(CFList)
print(len(CFList))
print(np.shape(CFData))
print()

WindPowerList=(list(map(lambda x:'Wind Power [MWh/year]-'+x,NameList2)))

PowerData=[]
for i in np.array(da['hgt']):
    for j in np.array(da['Turb']):  
        PowerData.append(np.array(da['Power'].sel(hgt=i,Turb=j)))

print(WindPowerList)
print(np.shape(PowerData))
```

```python
NormList=[]

for i in HGT:
    NameDir=list(map(lambda x: x+"-"+i,np.array(da['normData'])))
    NormList.extend(NameDir)    
print(NormList)

NormData=[]
for i in np.array(da['hgt']):
    for j in np.array(da['normData']):  
        NormData.append(np.array(da['Norn'].sel(hgt=i,normData=j)))
print(np.shape(NormData))
```
