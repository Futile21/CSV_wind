```python
# da=xr.open_dataset(r"C:\AllRun\Safrica000_Run10.nc")
da=xr.open_mfdataset(r"C:\AllRun\*10.nc",combine='by_coords')
da
```

```python
DATA_DIR = r"C:\AllRun"
filelist = os.listdir(DATA_DIR)
# filelist
```

```python
Names=["N","NNE","NE","ENE","E","ESE","SE","SSE","S","SSW","SW","WSW","W","WNW","NW","NNW"]
step=360/16
cut=np.arange(step,361,step)
cut=cut-step/2
```

```python
RoseData=[]


for ID in np.array(da['ID']):
    
# for ID in range(0,20):
    print(f'ID is {ID}')
    
    ######

    RosePoint = []

    for hgt in np.array(da['hgt']):
    #     print(f'hgt is {hgt}')
        WindDri = np.array(da['WDRO'].sel(hgt=hgt,ID=ID))
        windLen = len(WindDri)
        WindDri2 = sorted(WindDri)
        Rose = []

        for i in range(len(cut)):
    #         print(f'i is {i}')
            Rose.append(bisect.bisect_left(WindDri2, cut[i]))

        Rose = np.array(Rose)
        add =(windLen - Rose[-1])

        for j in range(len(Rose) - 2, -1, -1):
            Rose[j + 1] = Rose[j + 1] - Rose[j]

        Rose[0] = Rose[0] + add


        RosePoint.append(np.round(Rose / windLen * 100, 3))
        
    RoseData.append(RosePoint)

```
